## Simple Moving Average (SMA)

## Bollinger bands

```python
# Bollinger Channel High Band
df_copy['volatility_bbh'] = ta.volatility.bollinger_hband(df_copy['Close'], window=20, window_dev=2, fillna=False) # valeurs par dÃ©faut

# Bollinger Channel Indicator Crossing High Band (binary)
df_copy['volatility_bbhi'] = ta.volatility.bollinger_hband_indicator(df_copy['Close'], window=20, window_dev=2, fillna=False) # valeurs par dÃ©faut

# Bollinger Channel Low Band
df_copy['volatility_bbl'] = ta.volatility.bollinger_lband(df_copy['Close'], window=20, window_dev=2, fillna=False) # valeurs par dÃ©faut

# Bollinger Channel Indicator Crossing Low Band (binary)
df_copy['volatility_bbli'] = ta.volatility.bollinger_lband_indicator(df_copy['Close'], window=20, window_dev=2, fillna=False) # valeurs par dÃ©faut

# Bollinger Channel Middle Band
df_copy['volatility_bbm'] = ta.volatility.bollinger_mavg(df_copy['Close'], window=20, fillna=False) # valeurs par dÃ©faut

# Bollinger Channel Percentage Band
df_copy['volatility_bbp'] = ta.volatility.bollinger_pband(df_copy['Close'], window=20, window_dev=2, fillna=False) # valeurs par dÃ©faut

# Bollinger Channel Band Width
df_copy['volatility_bbw'] = ta.volatility.bollinger_wband(df_copy['Close'], window=20, window_dev=2, fillna=False) # valeurs par dÃ©faut

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_copy[["Close", "volatility_bbh", "volatility_bbhi", "volatility_bbl", "volatility_bbli", "volatility_bbm", "volatility_bbp", "volatility_bbw"]].plot(title="Bollinger Bands");
```

Pour la visualisation dans un mÃªme graphique : volatility_bbh (High BB), volatility_bbl (Low BB), volatility_bbm (EMA BB). Eventuellement ajoutez la courbe du Close Price.

## RSI

```python
df_copy['momentum_rsi'] = ta.momentum.RSIIndicator(df_copy['Close'], window=14, fillna=False).rsi()
df_copy['momentum_rsi']
```

```python
fig, ax = plt.subplots()
df_copy['momentum_rsi'].plot(ax=ax, 
               title="Relative Strength Index (RSI)")
ax.hlines(y=30, 
          xmin=df_copy.index.min(), 
          xmax=df_copy.index.max(), 
          color="red")
ax.hlines(y=70, 
          xmin=df_copy.index.min(), 
          xmax=df_copy.index.max(), 
          color="red")

sns.despine()
plt.tight_layout()
```

## Moving Average Convergence Divergence (MACD)

```python
df_copy['trend_macd'] = ta.trend.MACD(df_copy['Close'], window_slow=26, window_fast=12, window_sign=9, fillna=False).macd() # Valeurs par dÃ©faut
df_copy['trend_macd']
```

```python
df_copy['trend_macd_signal'] = ta.trend.MACD(df_copy['Close'], window_slow=26, window_fast=12, window_sign=9, fillna=False).macd_signal() # Valeurs par dÃ©faut
df_copy['trend_macd_signal']
```

```python
fig, ax = plt.subplots(2, 1, sharex=True)

df_copy[["trend_macd", "trend_macd_signal"]].plot(ax=ax[0], title="Moving Average Convergence Divergence (MACD)")

ax[1].bar(df_copy.index, df_copy["trend_macd_signal"].values, label="trend_macd_signal")
ax[1].legend()

sns.despine()
plt.tight_layout()
```
