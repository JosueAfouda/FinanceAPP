## Bollinger bands

```python
# Bollinger Channel High Band
df_copy['volatility_bbh'] = ta.volatility.bollinger_hband(df_copy['Close'], window=20, window_dev=2, fillna=False) # valeurs par défaut

# Bollinger Channel Indicator Crossing High Band (binary)
df_copy['volatility_bbhi'] = ta.volatility.bollinger_hband_indicator(df_copy['Close'], window=20, window_dev=2, fillna=False) # valeurs par défaut

# Bollinger Channel Low Band
df_copy['volatility_bbl'] = ta.volatility.bollinger_lband(df_copy['Close'], window=20, window_dev=2, fillna=False) # valeurs par défaut

# Bollinger Channel Indicator Crossing Low Band (binary)
df_copy['volatility_bbli'] = ta.volatility.bollinger_lband_indicator(df_copy['Close'], window=20, window_dev=2, fillna=False) # valeurs par défaut

# Bollinger Channel Middle Band
df_copy['volatility_bbm'] = ta.volatility.bollinger_mavg(df_copy['Close'], window=20, fillna=False) # valeurs par défaut

# Bollinger Channel Percentage Band
df_copy['volatility_bbp'] = ta.volatility.bollinger_pband(df_copy['Close'], window=20, window_dev=2, fillna=False) # valeurs par défaut

# Bollinger Channel Band Width
df_copy['volatility_bbw'] = ta.volatility.bollinger_wband(df_copy['Close'], window=20, window_dev=2, fillna=False) # valeurs par défaut

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

## Forecasting with Prophet

```python
import pandas as pd
import numpy as np
from prophet import Prophet
from prophet.plot import add_changepoints_to_plot
import yfinance as yf
```

```python

df = yf.download("BTC-USD", 
                 start="2019-01-01", 
                 end="2023-11-07",
                 progress=False,
                 interval="1d")
df
```

```python
df['Close'].plot(title="Daily Bitcoin prices (2019-2023)");
```

```python
btc_close = df[['Close']]
btc_close = btc_close.reset_index(drop=False)
btc_close.columns = ['ds', 'y']
btc_close
```

```python
btc_close.info()
```

```python
# Train/Test split
train_indices = btc_close['ds'] < "2022-11-07"
btc_train = btc_close[train_indices]
btc_test = btc_close[~train_indices].reset_index(drop=True)
```

On peut aussi utiliser cette méthode :

```python
from sklearn.model_selection import train_test_split

# Sequential train/test split with 80% for training and 20% for testing
btc_train2, btc_test2 = train_test_split(btc_close, test_size=0.2, shuffle=False, random_state=42)

# btc_train will contain 80% of the data, and btc_test will contain 20%.
```

```python
# Create and fit the model
prophet = Prophet(
    changepoint_range=0.9
)
prophet.add_country_holidays(country_name="US")
prophet.add_seasonality(name="annual", 
                        period=365, 
                        fourier_order=5)
prophet.fit(btc_train)
```

```python
# merge the test values with the forecasts
SELECTED_COLS = [
    "ds", "yhat", "yhat_lower", "yhat_upper"
]

btc_pred = (
    btc_pred
    .loc[:, SELECTED_COLS]
    .reset_index(drop=True)
)
btc_test = btc_test.merge(btc_pred, on=["ds"], how="left")
btc_test["ds"] = pd.to_datetime(btc_test["ds"])
btc_test = btc_test.set_index("ds")
btc_test.head()
```

```python
# Plot the test values VS predictions
fig, ax = plt.subplots(1, 1)

PLOT_COLS = [
    "y", "yhat", "yhat_lower", "yhat_upper"
]
ax = sns.lineplot(data=btc_test[PLOT_COLS])
ax.fill_between(
    btc_test.index,
    btc_test["yhat_lower"],
    btc_test["yhat_upper"],
    alpha=0.3
)

ax.set(
    title="Bitcoin Price - actual vs. predicted",
    xlabel="Date",
    ylabel="Bitcoin Price ($)"
)

plt.tight_layout()
sns.despine()
```

```python
import plotly.graph_objs as go

# Create a figure for the interactive chart
fig = go.Figure()

# Plot the actual values ('y')
fig.add_trace(go.Scatter(x=btc_test.index, y=btc_test['y'], mode='lines', name='Actual', line=dict(color='blue')))

# Plot the predicted values ('yhat')
fig.add_trace(go.Scatter(x=btc_test.index, y=btc_test['yhat'], mode='lines', name='Predicted', line=dict(color='orange')))

# Fill the region between the lower and upper bounds
fig.add_trace(go.Scatter(x=btc_test.index, y=btc_test['yhat_upper'], mode='lines', line=dict(width=0), showlegend=False, fill='tonexty', fillcolor='rgba(255,165,0,0.3)', name='Uncertainty'))

fig.add_trace(go.Scatter(x=btc_test.index, y=btc_test['yhat_lower'], mode='lines', line=dict(width=0), fill='tonexty', fillcolor='rgba(255,165,0,0.3)', showlegend=False))

# Customize the layout
fig.update_layout(
    title="Bitcoin Price - Actual vs. Predicted",
    xaxis_title="Date",
    yaxis_title="Bitcoin Price ($)",
)

# Show the interactive chart
#st.plotly_chart(fig)
fig
```
